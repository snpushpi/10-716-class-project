```python
import pandas as pd
import os

# Path to the massive CSV file
input_file = 'nasdaq_exteral_data.csv'

# Folder where split files will be saved
output_folder = 'news_data_raw'
os.makedirs(output_folder, exist_ok=True)

# Define a chunk size to avoid loading the whole file into memory
chunksize = 100000  # Adjust the chunk size based on your system's memory

# Process the CSV file in chunks
for chunk in pd.read_csv(input_file, chunksize=chunksize):
    # Group by the 'Stock_symbol' column
    for stock_symbol, group in chunk.groupby('Stock_symbol'):
        # Create a safe filename for the stock symbol
        output_file = os.path.join(output_folder, f"{stock_symbol}.csv")
        
        # If the file already exists, append without writing the header
        if os.path.exists(output_file):
            group.to_csv(output_file, mode='a', header=False, index=False)
        else:
            group.to_csv(output_file, mode='w', header=True, index=False)

print("CSV splitting complete.")
```

```python
import os
import csv

# Set your folder path and column name
folder_path = "stock_news_author_integrated/"  # Change this to your folder path
column_name = "Date"  # Change this to the name of the column you're interested in

# Iterate over files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                reader = csv.DictReader(f)  # Read as a dictionary
                column_values = set()  # Use a set to store unique values

                for row in reader:
                    if column_name in row:
                        column_values.add(row[column_name])  # Add the value to the set

                print(f"{filename}: {len(column_values)} unique values in column '{column_name}'")

        except Exception as e:
            print(f"Error processing {filename}: {e}")
```

```python
import csv
from collections import defaultdict

# Set your file path and column name
file_path = "/home/spushpit/FNSPID/stock_news_author_integrated/AAL.csv"  # Change this to your CSV file path
column_name = "Date"  # Change this to the column you're interested in

grouped_rows = defaultdict(list)

with open(file_path, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        if column_name in row:
            grouped_rows[row[column_name]].append(row)

# Filter to only keep entries with more than 1 row
multiple_rows_groups = {key: rows for key, rows in grouped_rows.items() if len(rows) > 1}

# Get the first 2 entries with multiple rows
first_two_multiple_rows = list(multiple_rows_groups.items())[:2]

# Print the first 10 rows for these two entries with multiple rows
for key, rows in first_two_multiple_rows:
    print(f"Entries for '{column_name}' = {key}:")
    for i, row in enumerate(rows[:10]):  # Limit to the first 10 rows
        print(f"Row {i + 1}: {row}")
    print("\n" + "-"*50 + "\n")
```
